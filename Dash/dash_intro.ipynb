# Example-1 HTML tags and Markdown

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

app.layout = html.Div([ 
    html.Br(),
    html.H1('Hello Dash!!!'),
    html.Div('Dash : A data product development framework from plotly'),
    dcc.Markdown(
'''
### Dash and Markdown

Dash supports [Markdown](https://commonmark.org/).

Markdown is a simple way to write and format text.
It includes a syntax for things like "bold text" and "italics",
[links](https://commonmark.org/help), inline `code snippers, lists, quotes and more.`
''' ,
        style = {'textAlign':'right',
                    'color':'#456FBA'}
    )
     
])

if __name__ == '__main__':
    app.run_server()
```

# Example-8

```python
import dash
import dash_renderer
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np

app = dash.Dash()

# np.random.seed(42)
# x_rand = np.random.randint(1,61,60)
# y_rand = np.random.randint(1,61,60)

x_rand = np.arange(-20,21,1)
y_rand = list(map(lambda x:2*x**2+2*x-250, x_rand))

app.layout = html.Div([
    dcc.Graph(
        id ='sample_chart',
        figure ={
            'data':[
                go.Scatter(
                    x = x_rand,
                    y = y_rand,
                    mode = 'markers'
                )
            ],
            'layout' : go.Layout(
                title = 'Scatterplot of Random 60 points',
                xaxis = {'title' : 'Random X Values'},
                yaxis = {'title' : 'Random Y Values'}
            )
        }
    )
])

if __name__ == '__main__':
    app.run_server(port=4050)
```

# Example-9

```python
import dash
import dash_renderer
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np
import pandas as pd

app = dash.Dash()

# pip install openpyxl
orders = pd.read_excel('sales.xlsx', engine='openpyxl')

# # pip install xlrd
# orders = pd.read_excel('sales.xls', engine='xlrd')

app.layout = html.Div([
    dcc.Graph(
        id ='sample_chart',
        figure ={
            'data':[
                go.Scatter(
                    x = orders.Sales,
                    y = orders.Profit,
                    mode = 'markers'
                )
            ],
            'layout' : go.Layout(
                title = 'Scatterplot of Random 60 points',
                xaxis = {'title' : 'Sales'},
                yaxis = {'title' : 'Profit'},
                hovermode = 'closest'
            )
        }
    )
])

if __name__ == '__main__':
    app.run_server(port=4050)
```

# Example-10 Dropdown

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

app.layout = html.Div([
    html.Label('Type of Strategy'),
    dcc.Dropdown(
            id='strategy',
            options=[
                {'label': 'Strategy-1', 'value': 'Strategy-1'},
                {'label': 'Strategy-2', 'value': 'Strategy-2'},
                {'label': 'Strategy-3', 'value': 'Strategy-3'},
                {'label': 'Strategy-4', 'value': 'Strategy-4'},
                {'label': 'Strategy-5', 'value': 'Strategy-5'}
            ], 
             placeholder='Select strategy...',
#              disabled = True         # shut down the selection
    ),
    
    html.Br(),
    html.Label('Choose a city'),    
    dcc.Dropdown(
        id = 'first-dropdown',
        options = [
            {'label' : 'San Francisco', 'value': 'SF'},
            {'label' : 'New York City', 'value': 'NYC'},
            {'label' : 'Washington', 'value': 'WA'},
            {'label' : 'Teksas', 'value': 'TE', 'disabled':True}, # make passive the 'Teksas' selection
            {'label' : 'Virginia', 'value': 'VA'},
            {'label' : 'New Jersey', 'value': 'NJ'}
            
        ],
        value = 'NYC',
        multi = True # enable to multiple selection
    )
])

if __name__ == '__main__':
    app.run_server()
```

# Example-11 Slider, RangeSlider, Input Box

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

app.layout = html.Div([
    html.Label('Slider - 1'),
    dcc.Slider(
        min = 1,
        max = 10,
        value = 5 # default starting value of slider
    ),
    
    html.Br(),
    
    html.Label('Slider - 2'),
    dcc.Slider(
        min = 1,
        max = 20,
        value = 5,
        step = .5, # default starting step of slider is 1
        marks = {i:i for i in range(1,21,2)}
    ),
    
    html.Br(),
    
    html.Label('RangeSlider'),
    dcc.RangeSlider(
        min = 1,
        max = 20,
        value = [4,10],
        step = .5, # default starting step of slider is 1
        marks = {i:i for i in range(21)}
    ),
    
    html.Br(),
    html.Br(),
  
    html.Div([
        html.Label('Input Box: '),
        dcc.Input(
            placeholder = 'Input Your Name...',
            type = 'text',
            value = ''
        ),
    ])
])

if __name__ == '__main__':
    app.run_server()
```

# Example-13 Check List, Radio List, DatePickerSingle

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from datetime import datetime as dt

app = dash.Dash()

app.layout = html.Div([
    html.Label('Check List: '),
    dcc.Checklist(
        options=[
            {'label': 'Montréal', 'value': 'MTL'},
            {'label' : 'San Francisco', 'value': 'SF'},
            {'label' : 'New York City', 'value': 'NYC'},
            {'label' : 'Washington', 'value': 'WA'},
            {'label' : 'Teksas', 'value': 'TE'}, 
            {'label' : 'Virginia', 'value': 'VA'},
            {'label' : 'New Jersey', 'value': 'NJ'}
        ],
        value=['NYC', 'NJ']
    ),
    
    html.Br(),
    html.Br(),
    html.Label('Radio List: '),
    dcc.RadioItems(
            options=[
                {'label': 'Positive', 'value': '+'},
                {'label': 'Negative', 'value': '-'},
                {'label': 'Zero', 'value': '0'},
           ],
            value='-'
    ),
    
    html.Br(),
    html.Br(),
    html.Label('Date Picker: '),
    dcc.DatePickerSingle(
        id = 'dt-pick-single',
        date = dt(2021,4,16)
    )
     
])

if __name__ == '__main__':
    app.run_server()
```

# Example-14 DatePickerSingle and DatePickerRange

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from datetime import datetime as dt

app = dash.Dash()

app.layout = html.Div([ 
    html.Br(),
    html.Label('Date Picker: '),
    dcc.DatePickerSingle(
        id = 'dt-pick-single',
        date = dt(2021,4,16)
    ),
    
    html.Br(),
    html.Br(),
    html.Br(),
    html.Label('Date Picker Range: '),
    dcc.DatePickerRange(
        id = 'dt-pick-range',
        start_date = dt(2021,4,16),
        end_date_placeholder_text = 'Select the end date...'
    )
     
])

if __name__ == '__main__':
    app.run_server()
```
